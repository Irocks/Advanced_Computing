```python
from sklearn.datasets import make_circles

X, y = make_circles(n_samples=400, factor=.3, noise=.1)
labels = ['b', 'r']
y = np.take(labels, (y < 0.5))

plt.scatter(X[:,0], X[:,1], c=y)
```

```python
#in order to collaps the data to simple logstic regression we will transform the data to polar coordinates
```

```python
R=np.sqrt(X[:,0]**2 + X[:,1]**2)
theta=np.arctan2(X[:,1], X[:,0])
```

```python
plt.scatter(theta, R , c=y)
```

```python
#Now the division is clear and we can fit the logistic regression
```
